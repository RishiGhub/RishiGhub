```python
for i in df.columns[:-1]:
    q1=df[i].quantile(.25)
    q3=df[i].quantile(.75)
    iq=q3-q1
    df.loc[(df[i]<q1-iq*1.5)|(df[i]>q3+iq*1.5),i]=df[i].median()
```

```python
from sklearn.model_selection import train_test_split
x=df.drop('Signal_Strength',axis=1)
xtrain,xtest,ytrain,ytest=train_test_split(x,df['Signal_Strength'],test_size=.3,random_state=0)
import random
random.seed(0)
import tensorflow.keras as k
m=k.models.Sequential([k.layers.BatchNormalization(input_shape=(11,)),k.layers.Dense(200,activation='relu'),k.layers.Dense(100,activation='relu'),k.layers.Dense(60,activation='relu'),k.layers.Dense(30,activation='relu'),k.layers.Dense(6,activation='softmax')])
```

```python
m.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
ytrain = ytrain-3
ytest=ytest-3
ytrain=k.utils.to_categorical(ytrain,num_classes=6)
ytest=k.utils.to_categorical(ytest,num_classes=6)
m.fit(xtrain,ytrain,epochs=1000,batch_size=64,validation_data=(xtest,ytest))
```
